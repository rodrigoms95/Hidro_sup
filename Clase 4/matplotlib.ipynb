```python
# Matplotlib es la principal librer√≠a para graficar en python.
# Tiene varias interfaces de trabajo, utilizaremos Pyplot.

# Tomado de: https://matplotlib.org/stable/tutorials/introductory/pyplot.html

import numpy as np
import pandas as pd

# Importamos pyplot
from matplotlib import pyplot as plt
```

```python
# Elementos de una gr√°fica.

# Una figura representa toda la gr√°fica en conjunto, un lienzo.
# Un eje es un dbujo, l√≠nea, etc. con sus ejes sobre el lienzo.

# Incializamos gr√°fica.
# Declarar nueva figura.
fig, ax = plt.subplots()
# Fondo blanco.
fig.set_facecolor("w")
# Tama√±o de la figura.
fig.set_size_inches(10, 4)
# Resoluci√≥n.
fig.set_dpi(150) 

# Mostrar la figura.
plt.show()
```

```python
# Gr√°fica b√°sica

# Incializamos gr√°fica.
fig, ax = plt.subplots()
fig.set_facecolor("w")
fig.set_size_inches(6, 6)
fig.set_dpi(150) 

# Declaramos variables.
x = np.array( [1, 2, 3, 4] )
y = x

# Graficamos.
ax.plot(x, y)

# Estilo.
# L√≠mites de la gr√°fica.
ax.set_xlim(1, 4)
ax.set_ylim(1, 5)
# Cantidad de marcas en cada eje.
ax.xaxis.set_major_locator( plt.MaxNLocator(3) )
# Marcas personalizadas.
ax.set_yticks([1, 3, 4])
# Malla.
ax.grid()
# T√≠tulo.
ax.set_title("Mi primera gr√°fica", size = 10)
# Nombre de ejes.
ax.set_xlabel("Eje x")
ax.set_ylabel("Eje y")
# Texto en la gr√°fica.
fig.text(0.2, 0.3, "Hola")
# Anotaciones
ax.annotate("punto", xy = [2, 2], xytext = [3, 4],
    arrowprops = {"width": 1, "color": "red"})

plt.show()
```

```python
# Estilos.

# Incializamos gr√°fica.
fig, ax = plt.subplots()
fig.set_facecolor("w")
fig.set_size_inches(6, 6)
fig.set_dpi(150) 

# Declaramos variables.
x = np.linspace(1, 10, 20)

# Graficamos.
ax.scatter(x, x, s = 10, color = "r", marker = "^", zorder = 1)
ax.scatter(x, x**2, s = 50, color = "g", marker = "s", zorder = 3)
ax.plot(x, x**2.5, color = "blue", linewidth = 5, alpha = 0.5, zorder = 2)

# Leyendas.
leyenda = ["tri√°ngulo", "cuadrado", "l√≠nea"]
ax.legend(leyenda)

plt.show()
```

```python
# Tener m√∫ltiples ejes.

# Incializamos gr√°fica.
fig, ax = plt.subplots(2)
fig.set_facecolor("w")
fig.set_size_inches(6, 6)
fig.set_dpi(150)

# Declaramos variables.
x = np.linspace(0, 10, 1000)

# graficamos en cada eje.
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))

# Estilo.
#for a in ax:
#    a.set_xlim(0, 10)
#    a.set_ylim(-1.5, 1.5)
ax[0].set_xlim(0, 10)
ax[0].set_ylim(-1.5, 1.5)
ax[1].set_xlim(0, 10)
ax[1].set_ylim(-1.5, 1.5)

# Separaci√≥n adecuada entre gr√°ficas cuando hay varios ejes.
fig.tight_layout()
plt.show()
```

```python
# Tener m√∫ltiples ejes.

# Incializamos gr√°fica.
fig, ax = plt.subplots(2, 2)
fig.set_facecolor("w")
fig.set_size_inches(6, 6)
fig.set_dpi(150)

# Declaramos variables.
x = np.linspace(0, 10, 1000)

# graficamos en cada eje.
ax[0, 0].plot(x, np.sin(x))
ax[1, 0].plot(x, np.cos(x))
ax[0, 1].plot(x, np.tan(x))
ax[1, 1].plot(x, np.tanh(x))

# Estilo.
#for a in ax.flatten():
#    a.set_xlim(0, 10)
#    a.set_ylim(-1.5, 1.5)
ax[0].set_xlim(0, 10)
ax[0].set_ylim(-1.5, 1.5)
ax[1].set_xlim(0, 10)
ax[1].set_ylim(-1.5, 1.5)
    
fig.tight_layout()
plt.show()
```

```python
# Ejes logar√≠tmicos.

# Incializamos gr√°fica.
fig, ax = plt.subplots()
fig.set_facecolor("w")
fig.set_size_inches(6, 6)
fig.set_dpi(150) 

# Declaramos variables.
x = np.linspace(1, 100, 10000)
y = np.exp(x)

# Graficamos.
ax.plot(x, y, linestyle = "dashed" )

# Estilo.
ax.set_yscale("log")

plt.show()
```

```python
# Gr√°fica polar.

# Incializamos gr√°fica.
fig, ax = plt.subplots(subplot_kw = {"projection": "polar"})
fig.set_facecolor("w")
fig.set_size_inches(6, 6)
fig.set_dpi(150) 

# Declaramos variables.
r = np.arange(0, 2, 0.01)
theta = 2 * np.pi * r

# Graficamos.
ax.plot(theta, r)

# Estilo.
ax.set_rmax(2)
ax.set_rticks([0.5, 1, 1.5, 2])
# Rotar posici√≥n de marcas.
ax.set_rlabel_position(-22.5)
ax.grid(True)

plt.show()
```

```python
# Esposible graficar directo desde pandas.

# Incializamos gr√°fica.
fig, ax = plt.subplots(4)
fig.set_facecolor("w")
fig.set_size_inches(6, 10)
fig.set_dpi(150) 

# Creamos un dataframe aleatorio.
df = pd.DataFrame(np.random.randn(100,2), columns = ["Col_1", "Col_2"])

print(df.head())

# Todas las columnas.
df.plot(ax = ax[0])
# Una columna contra otra.
df.plot("Col_1", "Col_2", ax = ax[1], kind = "scatter")
# Una solar columna.
df["Col_2"].plot(ax = ax[2], kind = "bar")
# Histograma.
df["Col_2"].hist(ax = ax[3], bins = 20, density = True)

fig.tight_layout()
plt.show()
```
